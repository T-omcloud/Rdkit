3.操作分子

```python
# 获取分子中的原子GetAtoms()
smi='CC(C)OC(=O)C(C)NP(=O)(OCC1C(C(C(O1)N2C=CC(=O)NC2=O)(C)F)O)OC3=CC=CC=C3'
mol = Chem.MolFromSmiles(smi)

# 获取分子中的所有原子，返回一个原子序列对象
atoms = mol.GetAtoms()

# 原子对象组成的列表
print(type(atoms))
# 获取第一个原子对象
print(type(atoms[0]))
```

```python
# 获取分子中的键 GetBonds()
smi='CC(C)OC(=O)C(C)NP(=O)(OCC1C(C(C(O1)N2C=CC(=O)NC2=O)(C)F)O)OC3=CC=CC=C3'
mol = Chem.MolFromSmiles(smi)

# 获取分子中的所有键，返回一个键序列对象
bonds = mol.GetBonds()

# 键对象组成的列表
print(bonds)
print(type(bonds))
# 获取第一个键对象
print(type(bonds[0]))
```

```python
# 根据原子编号获取键 GetAtomWithIdx()
smi='CC(C)OC(=O)C(C)NP(=O)(OCC1C(C(C(O1)N2C=CC(=O)NC2=O)(C)F)O)OC3=CC=CC=C3'
mol = Chem.MolFromSmiles(smi)

atom0 = mol.GetAtomWithIdx(0)
print(atom0.GetSymbol())  # 输出：C
print(type(atom0))
```

```python
# 获取原子的坐标信息 mol.GetConformer().GetAtomPosition(n)

# 1. 生成并构建3D分子
smi = 'CC(C)OC(=O)C(C)NP(=O)(OCC1C(C(C(O1)N2C=CC(=O)NC2=O)(C)F)O)OC3=CC=CC=C3'
mol = Chem.AddHs(Chem.MolFromSmiles(smi))  # 一步生成+加氢
AllChem.EmbedMolecule(mol)                 # 生成3D坐标
AllChem.UFFOptimizeMolecule(mol)           # 优化结构

# 2. 简化获取坐标
conf = mol.GetConformer()  

# 获取 1号原子坐标
x, y, z = conf.GetAtomPosition(1)  

# 获取单个坐标
x1 = conf.GetAtomPosition(1).x
y1 = conf.GetAtomPosition(1).y
z1 = conf.GetAtomPosition(1).z

# 获取 3号原子坐标（列表格式）
xyz3 = list(conf.GetAtomPosition(3))

# 获取所有原子坐标
print("序号\t元素\tx\ty\tz")
print("-"*50)

for idx in range(mol.GetNumAtoms()):
    atom = mol.GetAtomWithIdx(idx)       # 获取原子
    symb = atom.GetSymbol()              # 获取元素符号
    x, y, z = conf.GetAtomPosition(idx)   # 获取坐标
    print(f"{idx}\t{symb}\t{x:.2f}\t{y:.2f}\t{z:.2f}")
```

```python
# 根据键的编号获取键 GetBondWithIdx( )
# 注意：键的编号也是从 0 开始数
smi='CC(C)OC(=O)C(C)NP(=O)(OCC1C(C(C(O1)N2C=CC(=O)NC2=O)(C)F)O)OC3=CC=CC=C3'
mol = Chem.MolFromSmiles(smi)

# 获取第 1 号键（第二个键）
bond1 = mol.GetBondWithIdx(1)
print(type(bond1))

# 查看键的类型：单键/双键/三键
print(bond1.GetBondType())

# 查看这个键连接的两个原子
print(bond1.GetBeginAtomIdx(), bond1.GetEndAtomIdx())
```

5.处理2D分子：描述

```python
# 产生2D坐标 AllChem.Compute2DCoords(m)
# 该算法能够减少分子中原子在平面内的碰撞，使得绘制的分子更加清晰。对于复杂分子如青蒿素依旧不能解决平面碰撞问题。  
smis=[
    'COC1=C(C=CC(=C1)NS(=O)(=O)C)C2=CN=CN3C2=CC=C3',
#     'CCN(CC1=C(C=CC(=C1)C(F)(F)F)C2=CC(=C3N2C=NC=C3)CC(=O)O)C(=O)C4CC4',
    'C1=CC2=C(C(=C1)C3=CN=CN4C3=CC=C4)ON=C2C5=CC=C(C=C5)F',
    'COC(=O)C1=CC2=CC=CN2C=N1',
    'C1=C2C=C(N=CN2C(=C1)Cl)C(=O)O',
]

# 模板分子：用来统一骨架
# 所有分子都会按照这个模板的形状来画图，保证同一位置的原子在不同分子中都在同一位置上。
template = Chem.MolFromSmiles('c1nccc2n1ccc2')
AllChem.Compute2DCoords(template)
mols=[]

# 批量画图：一行4个，每个200x200
for smi in smis:
    mol = Chem.MolFromSmiles(smi)
    AllChem.GenerateDepictionMatching2DStructure(mol,template)
    mols.append(mol)
img=Draw.MolsToGridImage(mols,molsPerRow=4,subImgSize=(400,400),legends=['' for x in mols])
img
```

6.处理3D分子

RDKit中内置了2种不同的方法产生分子构象。 

方法一：距离几何法
 1. 基于链接表和规则计算分子中原子的距离矩阵； 
 2. 使用三角平滑算法对距离矩阵进行优化； 
 3. 产生一个随机距离矩阵； 
 4. 为每个原子产生一个坐标； 
 5. 使用粗糙的立场对坐标进行优化
通过该方法产生的3D构象通常很丑陋，需要进一步通过力场进行美化。RDKit美化分子构象使用的是UFF力场

方法二：基于CSD的扭转角偏好
通过该方法产生的构象不需要进一步通过力场进行优化。从2018.09版本开始，RDKit产生3D构象的默认方法是方法二

```python
# 产生3D构象 AllChem.EmbedMolecule(m)

# 产生3D构象前，需要为分子添加H原子。
m = Chem.MolFromSmiles('OC1C2C1CC2')
m1 = Chem.AddHs(m)

AllChem.EmbedMolecule(m1) 
Draw.ShowMol(m1, size=(600, 600))
```

```python
# 产生多个3D构象

smi='CCCC1CCCC(C(C)C)C1'
m2=Chem.MolFromSmiles(smi)

AllChem.EmbedMultipleConfs(m2, numConfs=20) 

Draw.ShowMol(m2, size=(600, 600)) #不加H的形式

writer = Chem.PDBWriter('1.pdb')  # 创建PDB写入器，文件名为1.pdb
writer.write(m2)                  # 把分子（含多个构象）写入文件
writer.close()                    # 关闭文件
```

```python
# MMFF力场对构象进行优化 AllChem.MMFFOptimizeMoleculeConfs
res = AllChem.MMFFOptimizeMoleculeConfs(m1)
Draw.ShowMol(m1, size=(600, 600))
```

9.子结构搜索    

```python
# 判断是否有子结构m.HasSubstructMatch(patt)
m = Chem.MolFromSmiles('c1ccccc1OC')
patt = Chem.MolFromSmarts('OC')
flag = m.HasSubstructMatch(patt)
if flag:
    print("molecu m contains group -OCH3")
else:
    print("molecu m don't contain group -OCH3")
```

```python
# 获取第一个子结构对应的原子编号m.GetSubstructMatch()
if flag:
    atomids = m.GetSubstructMatch(patt)
    print("matched atom id:",atomids)
    # 从输出我们知道6号原子对应的是甲氧基原子的O原子，7号原子对应的是甲氧基原子的C原子。
else:
    print("molecu m don't contain group -OCH3")


for atom in m.GetAtoms():
    atom.SetProp('atomLabel', str(atom.GetIdx()))
m
```

```python
# 获取对应所有子结构对相应的原子编号 m.GetSubstructMatches
m = Chem.MolFromSmiles('c1ccc(OC)cc1OC')
patt = Chem.MolFromSmarts('OC')
flag =m.HasSubstructMatch(patt)

if flag:
    atomids = m.GetSubstructMatches(patt)
    print("matched atom id:",atomids)
else:
    print("molecu m don't contain group -OCH3")

for atom in m.GetAtoms():
    atom.SetProp('atomLabel', str(atom.GetIdx()))
m
```
